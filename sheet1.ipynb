```python
def get_acquirer(row):
    acq_fields = [row['CAPTURE_ACQUIRER'], row['PURCHASE_ACQUIRER'], row['AUTH_ACQUIRER']]
    all_acq = ' '.join([str(a).lower() for a in acq_fields if pd.notna(a)])

    if any(x in all_acq for x in ['mastercard', 'firstdata', 'diners']):
        return 'FD'
    elif 'visa' in all_acq:
        return 'PRISMA'
    else:
        return next((a.upper() for a in acq_fields if pd.notna(a)), '')
```

```python
df['GTWT_ACQUIRER'] = df.apply(get_acquirer, axis=1)
```

```python
df
```

```python
def infer_brand(row):
    pay_method = str(row['PAY_METHOD']).upper()
    if any(k in pay_method for k in ['MASTER', 'MAESTRO', 'MASTERCARD']):
        return 'MASTERCARD'
    num_prefix = str(row['CARD_SIX_FIRST_DIGITS'])[:2]
    if num_prefix in ['34', '37']:
        return 'AMERICAN EXPRESS'
    elif num_prefix.startswith('5') or num_prefix.startswith('2'):
        return 'MASTERCARD'
    elif num_prefix.startswith('4'):
        return 'VISA'
    else:
        return 'PENDENTE'
```

```python
df['BRAND'] = df.apply(infer_brand, axis=1)
```

```python
df
```

```python
def get_first_non_empty(row):
    for col in ['PURCHASE_MERCHANT_NUMBER', 'CAPTURE_MERCHANT_NUMBER', 'AUTH_MERCHANT_NUMBER']:
        if pd.notna(row[col]) and row[col] != '':
            return row[col]
    return ''
```

```python
df['GTWT_MERCHANT_NUMBER'] = df.apply(get_first_non_empty, axis=1)
```
